```python
from bs4 import BeautifulSoup
import requests
```

```python
req = requests.get("https://charity.lovetoknow.com/List_of_Nonprofit_Organizations?fbclid=IwAR13LiN-E2_T5TTtji63f8XGWiu3-h7dOAmAKcQ6d9Y_1hY9H-uBYN8ngwk")
```

```python
soup = BeautifulSoup(req.content, 'html.parser')
```

```python
print (soup.prettify()[:])
```

```python
def find_categories (ssoup : BeautifulSoup):
    mycats = ssoup.find_all('h2')
    ans = []
    for name in mycats:
        ans.append(name.text)
    return ans
```

```python
category_list = find_categories(soup)
```

```python
print(category_list)
```

```python
def find_charities (ssoup: BeautifulSoup):
    
    ans = []
    all_charity_list = ssoup.find_all('ul')
    for index, charity_list in enumerate(all_charity_list):
        if index != 1 and index != 5:
            temp = []
            charity_url = charity_list.find_all('a', href = True)
            for url in charity_url:
                if url.string != None:
                    temp.append((url['href'], url.string))
            ans.append(temp)
    return ans
        
#     charity_url = charity_list.find_all('a', href = True)
#     for url in charity_url:
#         temp.append((url['href'], url.string))
#     return temp

#         result.extend(th.find_all(text='href'))
#     charity_url = charity_list.find_all('a', href = True)
#     print(type(charity_url))
#     for i in names:
#         charity.append(charity_url['href'])
#     return zip(names, charity_url)

find_charities(soup)  
```

```python
def parse_charities (ssoup: BeautifulSoup):
    ans = dict()
    categories = find_categories(ssoup)
    charities = find_charities(ssoup)
    for counter, value in enumerate(categories):
        ans[value] = charities[counter]
    return ans
```

```python
my_dict = parse_charities(soup)
```
